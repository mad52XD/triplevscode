```python
import matplotlib.pyplot as plt
import pandas as pd
a = list(range(1, 6))
b=list(range(9, 14)          )
print(a)
print(b)
df_scatter = pd.DataFrame({'x': a, 'y': b})
df_scatter.plot.scatter(x='x', y='y')   
plt.title('Scatter Plot Example')
plt.xlabel('X-axis Label')  
plt.ylabel('Y-axis Label')
plt.show()
```
